```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python

# Create a DataFrame showing daily temperatures for several cities
df = pd.DataFrame({
    'City': ['Istanbul', 'Ankara', 'Gaziantep', 'Izmir', 'Bursa', 'Antalya', 'Konya', 'Adana'],
    'Day1': [25, 28, 32, 27, 26, 30, 29, 33],
    'Day2': [24, 27, 31, 26, 25, 30, 28, 32],
    'Day3': [26, 29, 33, 28, 27, 31, 30, 34],
    'Day4': [25, 28, 32, 27, 26, 29, 29, 33],
    'Day5': [27, 30, 34, 29, 28, 32, 31, 35],
    'Day6': [28, 31, 35, 30, 29, 33, 32, 36],
    'Day7': [29, 32, 36, 31, 30, 34, 33, 37],
})

# Display the DataFrame
print(df)
```

```python
df1 =  df.melt(id_vars=['City'], var_name = 'days',value_name='temperature')
df2 = df1.head(2)
df2
```

```python
df1 =  df.melt(id_vars=['City'], var_name = 'days',value_name='temperature',ignore_index=(True)).reset_index(drop=True)
df1
```

```python
n1=df1.groupby('City')['temperature'].mean().plot(kind='bar')
plt.title('Average Temperature by City')
plt.show()
```
